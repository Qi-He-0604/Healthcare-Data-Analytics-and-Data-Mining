```python
#calculate number_of_MRI for each state
number_of_MRI=pd.DataFrame(healthcare_q4[["Provider License Number State Code_1"]].groupby("Provider License Number State Code_1").size())
```

```python
#rename the coulmn
number_of_MRI=number_of_MRI.rename(columns={0:'number_of_MRI'})
```

```python
number_of_MRI.reset_index(inplace=True)
```

```python
number_of_MRI=number_of_MRI.rename(columns={"Provider License Number State Code_1":'State Code'})
number_of_MRI
```

```python
#import population table
path=r"C:\Users\jiaji\Downloads\population.csv"
population=pd.read_csv(path)
```

```python
#import state_name table
path=r"C:\Users\jiaji\Desktop\state_name.csv"
state=pd.read_csv(path)
```

```python
#combine state name and state code
state=state.rename(columns={'State Name':'Geography'})
```

```python
state
```

```python
#merge tables tagether
q4=(population.merge(state,on="Geography")).merge(number_of_MRI,on="State Code")
```

```python
q4.head()
```

```python
q4["population"]=q4["population"].str.replace(",","")
```

```python
q4["population"]=q4["population"].astype("float")
```

```python
#calculate the number of MRI clinics per 1000,000 population for each state
q4["share"]=q4["number_of_MRI"]/q4["population"]/1000000
```

```python
q4
```

```python
q4_f=q4[["State Code","share"]]
```

```python
q4_f.to_csv("q4.csv")
```
