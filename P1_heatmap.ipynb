```python
#read data
mri = pd.read_csv("healthcare_q4.csv")
population = pd.read_excel("Population By State US Census Bureau.xlsx")
#convert state name to abbreviation
population['Provider License Number State Code_1'] = population['Geography'].map(us_state_abbrev)
```

```python
#count number of MRI in each state
mri_count = mri.groupby("Provider License Number State Code_1").count()
mri_count = pd.merge(mri_count, population, on='Provider License Number State Code_1', how='left')
#calculate density
mri_density = pd.concat([
    mri_count['Provider License Number State Code_1'], 
    mri_count['Entity Type Code']*1000000/mri_count['Estimate; Total']
    ], axis=1)
mri_density = mri_density.rename(columns={0: "density"})
mri_density
```
